Etape 3:Filtrage du dataframe selon la serie country

```python
#identification les valeurs contenues dans la série countr avec l'argument unique()
df.country.unique()
```

```python
#création du tuple country_selection qui va servir au filtre
country_selection=('United States', 'Morocco')
#déclaration de la variable country_filter
country_filter = df.country.isin(country_selection)
```

```python
df = df[country_filter]
df
```

Etape 4: appliquer le filtre sur la serie price_paid dans le datafame df

```python
#Apporter des modifications à la serie price_paid en supprimant le symbole "$"
#faire une copie df_test du dataframe d'origine df
df_test=df.copy()
#remplacer le symbole "$" par le vide
df_test.price_paid = df_test.price_paid.apply(lambda x: x.replace("$",""))
#changer le type de données stockées dans la série
df_test.price_paid = df_test.price_paid.astype(float)
#appliquer le filtre 
df_test[df_test["price_paid"] >= 5.0]
```
